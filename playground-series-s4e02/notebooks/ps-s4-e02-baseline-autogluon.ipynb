```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
#global variables

SEED = 42
TARGET = 'NObeyesdad'
FILEPATH = '../data/'
```

```python
train_df =pd.read_csv(f"{FILEPATH}train.csv")
test_df = pd.read_csv(f"{FILEPATH}test.csv")

test_ids = test_df["id"]
test_df = test_df.drop(["id"], axis = 1)
train_df = train_df.drop(["id"], axis = 1)

train_df.head()
```

```python
train = TabularDataset(train_df)
test = TabularDataset(test_df)
time_limit = 3600
metric = 'accuracy'

predictor = TabularPredictor(label=TARGET, eval_metric=metric).fit(train, time_limit=time_limit, presets='best_quality')
```

```python
predictor.leaderboard()
```

```python
predictions = predictor.predict(test)
```

```python
submission = pd.DataFrame()
submission["id"] = test_ids
submission[TARGET] = predictions

# write submission dataframe to .csv file 
submission.to_csv('submission.csv', index = False)
submission.head()
```
